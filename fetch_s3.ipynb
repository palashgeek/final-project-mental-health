```python
import boto3
import os

# Initialize S3 client
s3 = boto3.client('s3')

# Specify the bucket name
bucket_name = 'mental-health-final-project'

# Create a directory to store downloaded files
if not os.path.exists('s3_data'):
    os.makedirs('s3_data')

# List objects in the bucket
response = s3.list_objects_v2(Bucket=bucket_name)

# Check if objects are present in the response
if 'Contents' in response:
    for obj in response['Contents']:
        # Fetch object data and save it locally
        s3.download_file(bucket_name, obj['Key'], f"s3_data/{obj['Key']}")
        print(f"Downloaded: {obj['Key']}")
else:
    print("Bucket is empty.")
```

```python
# Directory where JSON files are stored
json_dir = 's3_data'

# List to store Pandas DataFrames
pandas_dfs = []

# Iterate over each JSON file in the directory
for file_name in os.listdir(json_dir):
    if file_name.endswith('.json'):
        # Read JSON file into a Pandas DataFrame
        with open(os.path.join(json_dir, file_name), 'r') as file:
            json_data = json.load(file)
            df = pd.DataFrame(json_data)
            pandas_dfs.append(df)

# Concatenate all Pandas DataFrames into a single DataFrame
combined_df = pd.concat(pandas_dfs, ignore_index=True)

# Now you have a single Pandas DataFrame containing data from all JSON files
print(combined_df.head())
```

```python
import pandas as pd
import json
import os

# Directory where JSON files are stored
json_dir = 's3_data'

# List to store Pandas DataFrames
pandas_dfs = []

# Iterate over each JSON file in the directory
for file_name in os.listdir(json_dir):
    if file_name.endswith('.json'):
        # Read JSON file into a Pandas DataFrame
        with open(os.path.join(json_dir, file_name), 'r') as file:
            json_data = json.load(file)
            df = pd.DataFrame(json_data)
            pandas_dfs.append(df)

# Concatenate all Pandas DataFrames into a single DataFrame
combined_df = pd.concat(pandas_dfs, ignore_index=True)

# Now you have a single Pandas DataFrame containing data from all JSON files
print(combined_df.head())
```

```python
import json
import boto3
from decimal import Decimal

def lambda_handler(event, context):
    s3_bucket = 'mental-health-final-project'
    json_file_name = event['Records'][0]['s3']['object']['key']
    print(s3_bucket)
    print(json_file_name)
    
    
    s3_client = boto3.client('s3')
    dynamodb_client = boto3.resource('dynamodb')
    dynamodb_table = dynamodb_client.Table('mental_health')

    try:
        # Get the JSON object from S3
        response = s3_client.get_object(Bucket=s3_bucket, Key=json_file_name)
        json_data = json.loads(response['Body'].read().decode('utf-8'), parse_float=Decimal)
        
        # Insert the JSON data into DynamoDB
        with dynamodb_table.batch_writer() as batch:
            for item in json_data:
                batch.put_item(Item=item)
                
        return {
            'statusCode': 200,
            'body': 'Data successfully loaded into DynamoDB.'
        }
    except Exception as e:
        return {
            'statusCode': 500,
            'body': f'Error: {str(e)}'
        }
```
